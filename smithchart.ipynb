```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scattersmith(
    imag=[0.2],
    real=[0.4],
    marker_symbol='x',
    marker_size=8,
    marker_color="green",
    subplot="smith1"
))



fig.update_layout(
    smith=dict(
        realaxis_gridcolor='red',
        imaginaryaxis_gridcolor='blue',
        domain=dict(x=[0,0.5])
    )
)

fig.update_smiths(bgcolor="white")

fig.show()
```

```python
import matplotlib.pyplot as plt

 
circle1 = plt.Circle((0, 0), 3, color='k', fill=False)

fig, ax = plt.subplots()

plt.xlim(-5,5)
plt.xlabel('AU')
plt.ylim(-5,5)
plt.ylabel('AU')

plt.grid(linestyle='-')

ax.set_aspect(1)

ax.add_artist(circle1)

plt.title('Habitable Zone Around the Sun ', fontsize=8)

plt.show()
```
