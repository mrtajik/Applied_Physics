```python
#2.1
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline    

t = np.arange(0,4*np.pi,0.01)   # start,stop,step
w=1
E =10.7*np.cos(w*t-0.1239*np.pi)
plt.plot(t,E)
plt.title('E vs t')
plt.xlabel('t-time')
plt.ylabel('E')
plt.show()
```
