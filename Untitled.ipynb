```python
from sympy import *
import numpy as np
t=symbols('t')
# Heaviside defaults to "switch on" at 0. Use shifts to get the right one.
f=Heaviside(t-1) 
x = np.linspace(0,10,100)

 

plot(f,(t,0,1),x-2, np.exp(-x))
 
```

```python
import matplotlib.pyplot as plt
x1 = np.linspace(0.01,2,100)
 
x = [0,0]
y = [0, 10]

plt.step(x, y)
plt.plot(x1,1/(np.sqrt(x1)))
plt.show()
```
